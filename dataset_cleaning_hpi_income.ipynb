The HPI and Median Income datasets will require cleaning prior to being able to properly compare the data. I will also be dropping data that isn't relevant to the state by state analysis.

Drop rows containing "manufactured", "distress-free, and "developmental". 
Verify "traditional" for "hpi_type" are the only remaining rows

```python
hpi_df = hpi_df[~hpi_df['hpi_type'].isin(['distress-free', 'developmental', 'manufactured'])]
hpi_df['hpi_type'].unique()
```

"frequency" is the next column for review. I will only need the rows containing "quarterly" as the "frequency"

Determine all values present in the "frequency" column.

```python
print(hpi_df['frequency'].unique().tolist())
```

Drop rows containing "monthly". 
Verify "quarterly" for "frequency" are the only remaining rows

```python
hpi_df = hpi_df[~hpi_df['frequency'].isin(['monthly'])]
hpi_df['frequency'].unique()
```

"level" is the next column for review. I will only need the rows containing "state" and "USA or Census Division" in the "level" column

Drop rows containing "MSA"

I will later be comparing individual states to the United States as a whole. This data also contains the information for divisions of the country which I do not need for the projects purposes. The divisions can be found in the "place_name" column. I will be removing all divisions where the value in the "level" column equals "USA or Census Division" and the value in the "place_name" column is something other than "United States"  


Determine divisions where "level" equals "USA or Census Division" 

```python
hpi_df.loc[hpi_df['level'].isin(['USA or Census Division']),'place_name'].unique()
```

Drop all divisions above except for "United States". Since I am only removing entries from the 'place_name' column that are associated with 'USA or Census Division' in the 'level' column I should be left with 51 unique values in the 'place_name' column

```python
hpi_df = hpi_df[~hpi_df['place_name'].isin(['East North Central Division', 'East South Central Division', 'Middle Atlantic Division', 'Mountain Division', 'New England Division', 'Pacific Division', 'South Atlantic Division','West North Central Division', 'West South Central Division'])]
hpi_df['place_name'].unique()
```

I see that there are 52 entries for uniques values in the 'place_name' column. District of Columbia has been included. I will need to drop that as my Median Income dataset does not include data for DC

Drop District of Columbia from 'place_name" column and verify only 51 entries remain (all states and United States)

```python
hpi_df = hpi_df[~hpi_df['place_name'].isin(['District of Columbia',])]
hpi_df['place_name'].unique()
```

The median income dataset provided by the Census Bureau only contains yearly data. However, the HPI data for 'traditional' and 'prchase-only' is provided on a quarterly basis. I will be dropping quarters 2, 3, and 4 from this dataset and rely on quarter 4 data for comparing hpi to median income

Drop '2', '3', and '4' from the 'period' column and verify

```python
hpi_df = hpi_df[~hpi_df['period'].isin([4, 2, 3])]
hpi_df['period'].unique()
```

The median income dateset only provides data up to 2024. However, the HPI dataset provides data up to 2025. I will need to remove rows where the 'yr' equals '2025

Drop '2025' from 'yr' colum and verify only 1991-2024 are present

```python
hpi_df = hpi_df[~hpi_df['yr'].isin([2025])]
hpi_df['yr'].unique()
```

Import Median Income dataset

```python
med_inc_df = pd.read_csv('data/h08.csv')
```

```python
print(med_inc_df.columns.tolist())
```

The HPI and Median Income datasets are not formatted similarly. I will need to change the formatting of the Median Income dataset to match that of the HPI dataset. **I did use Chat GPT for this portion**

```python
# Melt to long format
df_melted = med_inc_df.melt(
    id_vars='State',
    var_name='variable',
    value_name='value'
)

# Extract year and type (Median income vs Standard error)
df_melted[['year', 'type']] = df_melted['variable'].str.extract(
    r'(\d{4})\s+(Median income|Standard error)'
)

# Drop rows that didn't match (important cleanup)
df_melted = df_melted.dropna(subset=['year', 'type'])

# Convert year to integer
df_melted['year'] = df_melted['year'].astype(int)

# Remove duplicates (optional but safe)
df_melted = df_melted.drop_duplicates(subset=['State', 'year', 'type'])

# Pivot into final format (SAFE version)
df_long = df_melted.pivot_table(
    index=['State', 'year'],
    columns='type',
    values='value',
    aggfunc='first'
).reset_index()

# Rename columns
df_long = df_long.rename(columns={
    'Median income': 'median_income',
    'Standard error': 'standard_error'
})

# Clean column names (remove pivot column name label)
df_long.columns.name = None

med_inc_df = df_long

# Final check
print(med_inc_df.head())
print(med_inc_df.shape)
print(med_inc_df['year'].unique())
```

I need to change the names of the 'State" and 'year' columns to match the HPI columns then verify

```python
med_inc_df.rename(columns={'State': 'place_name', 'year': 'yr'}, inplace=True)
print(med_inc_df.head())
```
