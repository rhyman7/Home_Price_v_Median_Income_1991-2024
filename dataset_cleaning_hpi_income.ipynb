The HPI and Median Income datasets will require cleaning prior to being able to properly compare the data. I will also be dropping data that isn't relevant to the state by state analysis.

```python
print(hpi_df['frequency'].unique().tolist())
```

The median income dataset provided by the Census Bureau only contains yearly data. However, the HPI data for 'traditional' and 'prchase-only' is provided on a quarterly basis. I will be dropping quarters 2, 3, and 4 from this dataset and rely on quarter 4 data for comparing hpi to median income

Drop '2', '3', and '4' from the 'period' column and verify

```python
hpi_df = hpi_df[~hpi_df['period'].isin([4, 2, 3])]
hpi_df['period'].unique()
```

The median income dateset only provides data up to 2024. However, the HPI dataset provides data up to 2025. I will need to remove rows where the 'yr' equals '2025

Drop '2025' from 'yr' colum and verify only 1991-2024 are present

```python
hpi_df = hpi_df[~hpi_df['yr'].isin([2025])]
hpi_df['yr'].unique()
```

Import Median Income dataset

```python
med_inc_df = pd.read_csv('data/h08.csv')
```

```python
print(med_inc_df.columns.tolist())
```

The HPI and Median Income datasets are not formatted similarly. I will need to change the formatting of the Median Income dataset to match that of the HPI dataset. **I did use Chat GPT for this portion**

```python

df_melted = med_inc_df.melt(
    id_vars='State',
    var_name='variable',
    value_name='value'
)


df_melted[['year', 'type']] = df_melted['variable'].str.extract(
    r'(\d{4})\s+(Median income|Standard error)'
)


df_melted = df_melted.dropna(subset=['year', 'type'])


df_melted['year'] = df_melted['year'].astype(int)


df_melted = df_melted.drop_duplicates(subset=['State', 'year', 'type'])


df_long = df_melted.pivot_table(
    index=['State', 'year'],
    columns='type',
    values='value',
    aggfunc='first'
).reset_index()


df_long = df_long.rename(columns={
    'Median income': 'median_income',
    'Standard error': 'standard_error'
})


df_long.columns.name = None

med_inc_df = df_long


print(med_inc_df.head())
print(med_inc_df.shape)
print(med_inc_df['year'].unique())
```

I need to change the names of the 'State" and 'year' columns to match the HPI columns then verify

```python
med_inc_df.rename(columns={'State': 'place_name', 'year': 'yr'}, inplace=True)
print(med_inc_df.head())
```

I found that 'Dictrict of Columbia', 'United States', and 'State' are present. Dropping those three to only go forward with the data for the 50 states

```python
med_inc_df = med_inc_df[~med_inc_df['place_name'].isin(['District of Columbia', 'United States', 'State'])]
med_inc_df['place_name'].unique()
```
